## Import our test dataset and scale it

```python
test_df = pd.read_csv("../data/test_df.csv", sep = ",")
test_df = test_df.drop(['wind_direct'], axis = 1)
```

```python
scaler = MinMaxScaler(feature_range = (0,1))
def data_scaler(df):
  station = df.copy()
  scaled = scaler.fit_transform(station[station.columns[:14]])
  scaled = pd.DataFrame(scaled, columns=station.columns[:14],index = station.index)
  for feature in scaled.columns:
    station[feature] = scaled[feature]
  return station
```

```python
test_df = data_scaler(test_df)
test_df
```

## Declare and load three models:
 - 6 hours output
 - 9 hours output
 - 12 hours output

```python
encoder_input_size = len(test_df.columns)
decoder_input_size = 1
output_size = 1
hidden_size = 20
drop_out = 0


# 6 hours model parameters
input_length = 72
output_length = 6
num_layers = 2
six_hrs_model = EncoderDecoder(device = device,encoder_input_size = encoder_input_size, input_length = input_length, output_length = output_length,
                       output_size = output_size, hidden_size = hidden_size, num_layers = num_layers, dropout = drop_out).to(device)
load_checkpoint('models/6h-model.tar', six_hrs_model)

# 9 hours model parameters
input_length = 24
output_length = 9
num_layers = 2
nine_hrs_model = EncoderDecoder(device = device,encoder_input_size = encoder_input_size, input_length = input_length, output_length = output_length,
                       output_size = output_size, hidden_size = hidden_size, num_layers = num_layers, dropout = drop_out).to(device)
load_checkpoint('models/9h-model.tar', nine_hrs_model)

# 12 hours model parameters
input_length = 24
output_length = 12
num_layers = 2
twelve_hrs_model = EncoderDecoder(device = device,encoder_input_size = encoder_input_size, input_length = input_length, output_length = output_length,
                       output_size = output_size, hidden_size = hidden_size, num_layers = num_layers, dropout = drop_out).to(device)
load_checkpoint('models/12h-model.tar', twelve_hrs_model)
```

```python
time_steps = [i for i in range(1,13)]
losses = []
for time_step in time_steps:
  (mse, _, _), _  = time_step_performance(time_step, twelve_hrs_model, test_df, scaler)
  losses.append(mse)

plt.figure(figsize=(20,8))

plt.plot(losses, marker = '+', markersize = 15)
plt.xticks([time_step - 1 for time_step in time_steps], time_steps)
plt.suptitle ("The loss at each future time steps of our Encoder Decoder model (with 12 output time steps)".format(time_step) , weight = 'bold' ,fontsize = 20  )
plt.xlabel('Time Steps (hours)', fontsize = 15)
plt.ylabel('MSE', fontsize = 15)
```
