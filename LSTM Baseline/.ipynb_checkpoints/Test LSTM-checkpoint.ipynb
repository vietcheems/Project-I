## Import our test dataset and scale it

```python
test_df = pd.read_csv("../data/test_df.csv", sep = ",")
test_df = test_df.drop(['wind_direct'], axis = 1)
```

```python
scaler = MinMaxScaler(feature_range = (0,1))
def data_scaler(df):
  station = df.copy()
  scaled = scaler.fit_transform(station[station.columns[:14]])
  scaled = pd.DataFrame(scaled, columns=station.columns[:14],index = station.index)
  for feature in scaled.columns:
    station[feature] = scaled[feature]
  return station
```

```python
test_df = data_scaler(test_df)
test_df
```

## Declare and load the best LSTM model

```python
input_size = len(test_df.columns)
hidden_size = 20
input_length = 24
num_layers = 2

lstm = LSTM(input_length = input_length, input_size = input_size,  hidden_size= hidden_size, 
            num_layers = num_layers, device = device).to(device)

load_checkpoint('model/lstm-model.tar', lstm)
```

```python
predictions = predict_model(lstm, test_df, scaler = scaler, print_loss = True)
```
